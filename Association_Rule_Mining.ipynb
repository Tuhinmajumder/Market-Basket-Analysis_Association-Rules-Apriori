```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
import pandas as pd
```

```python
aa=apriori(whiskey,min_support=0.1,use_colnames=True)
```

```python
rules=association_rules(aa,min_threshold=0.6)
```

```python
store_data['cnt']=1
```

```python
store_data.groupby(["CASH_MEMO","PRODUCT"])["cnt"].sum()
```

```python
store_data_1=store_data.groupby(["CASH_MEMO","PRODUCT"])["cnt"].sum().unstack()
```

```python
def make_boolean(x):
    if x>0:
        return 1
    else:
        return 0
```

```python
store_data_final=store_data_1.applymap(make_boolean)
```

```python
bb=apriori(store_data_final,min_support=0.001)
```

```python
rules_store=association_rules(bb,min_threshold=0.8)
```
